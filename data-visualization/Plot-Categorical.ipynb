```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Categorical Plot dengan Matplotlib

### Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.bar(langs,students)
plt.show()
```

```python
data = [
        [30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19]
       ]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X+0.00, data[0], color='b', width=0.25)
ax.bar(X+0.25, data[1], color='g', width=0.25)
ax.bar(X+0.50, data[2], color='r', width=0.25)
```

# Categorical Plot dengan Seaborn

## Barplots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

Kita juga dapat menambahkan dimensi pada data yang divisualisasikan dengan bar plot, yaitu dengan cara menambahkan parameter `hue`.

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
```

Menambahkan gradasi warna dengan parameter `palette`. Lebih lengkap mengenai `palette` pada seaborn dapat langsung mengunjungi [API Docs](https://seaborn.pydata.org/tutorial/color_palettes.html) resmi Seaborn.

```python
sns.barplot(x='size', y='total_bill', data=tips, palette='Blues_d')
```

## Categorical scatterplots

```python
# Load Dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

Parameter `jitter` untuk mengendalikan kerenggangan data. Jika `jitter=False` maka scatter plot akan lebih berhimpitan.

```python
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

Pendekatan selanjutnya adalah menggunakan parameter `kind` yang di-set ke nilai `swarm`. Parameter ini akan membuat representasi data lebih baik, namun hanya cocok untuk jumlah data yang relatif sedikit.

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

Kita juga dapat menambahkan dimensi pada data yang divisualisasikan dengan scatter plot, yaitu dengan cara menambahkan parameter `hue` pada method `catplot()`.

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

### Bar Plot dengan Matplotlib

Di sini akan ditampilkan distribusi untuk atribut **Type 1** pada dataset. Library yang digunakan adalah Matplotlib.

Pertama, kita harus memfilter data terlebih dahulu, yaitu dengan mengambil nilai yang unik dari atribut **Type 1**.

```python
type_1 = df['Type 1'].unique() # Ambil nilai yang unik
type_1 = np.sort(type_1) # Urutkan data
```

Kemudian hitung kemunculan nilai-nilai pada atribut **Type 1** kemudian hasilnya langsung diurutkan berdasarkan abjadnya.

```python
type_1_num = df['Type 1'].value_counts(ascending=True)
type_1_num = pd.DataFrame(type_1_num) # Ubah struktur data type_1_num ke dalam DataFrame
type_1_num = type_1_num.reset_index() # Reset index DataFrame
type_1_num = type_1_num.sort_values('index', ascending=True) # Urutkan berdasarkan Nama
type_1_num.head()
```

Pada tabel di atas terlihat bahwa nama atribut dari DataFramenya masih tidak sesuai, sehingga bisa kita sesuaikan terlebih dahulu menjadi nama atribut sebagai mana mestinya.

```python
type_1_num.columns = ['Type 1', 'Jumlah']
type_1_num.head()
```

Setelah terbentuk struktur data yang tepat, maka kita bisa langsung membuat visualisasinya dengan Matplotlib.

```python
# Bar plot dengan Matplotlib
plt.bar(type_1_num['Type 1'], type_1_num['Jumlah'])
plt.xticks(rotation='vertical') # posisi teks sumbu x
plt.show()
```

### Barplot dengan Pandas

Library Pandas memiliki metode visualisasinya tersendiri, namun sama seperti Seaborn, kelas visualisasi pada Pandas juga dibuat berdasarkan kelas visualisasi pada Matplotlib.

Lihat perbedaan jumlah kode program yang harus ditulis jika membuat plot dengan library Pandas. Visualisasi pada Pandas diimplementasikan dengan memanggil method `plot()` pada DataFrame yang akan divisualisasikan, dengan tambahan parameter `bar` untuk menentukan bentuk visualisasi yang akan dibuat.

```python
# Bar plot pada library Pandas
df['Type 1'].value_counts().plot(kind='bar')
plt.show()
```

Misalnya sekarang kita akan memvisualisasikan atribut **Generation**.

```python
generation = df['Generation'].unique()
generation_num = df['Generation'].value_counts()
```

```python
# Bar plot dengan Matplotlib
plt.bar(generation, generation_num)
plt.show()
```

```python
# Bar plot pada library Pandas
df['Generation'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

### Bar Plot Lebih dari Dua Dimensi

Sebelum membuat plot dengan lebih dari dua dimensi kita perlu memisahkan terlebih dahulu atribut mana saja yang akan divisualisasikan. Misalnya di sini kita akan menggunakan atribut **Total**, **HP**, dan **Generation**.

```python
df_total_hp = df[['Total', 'HP', 'Generation']]
df_total_hp.head()
```

Plot dataframe `df_total_hp` dengan visualisasi dari library Pandas.

```python
df_total_hp.plot.scatter('Total', 'HP', c='Generation', colormap='viridis')
```

Plot dataframe `df_total_hp` dengan visualisasi dari library Seaborn.

```python
sns.scatterplot(x='Total', y='HP', hue='Generation', data=df)
```

```python
sns.catplot(x='Generation', y='Total', hue='Legendary', kind='swarm', data=df)
```
